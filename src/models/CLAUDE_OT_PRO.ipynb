```python

import os
import requests
from dotenv import load_dotenv

# Load API key from .env file
load_dotenv()
api_key = os.getenv("CLAUDE_API_KEY")

if not api_key:
    raise ValueError("Missing CLAUDE_API_KEY in .env file")

# Define endpoint and headers
url = "https://api.anthropic.com/v1/messages"
headers = {
    "x-api-key": api_key,
    "anthropic-version": "2023-06-01",
    "content-type": "application/json"
}
```

```python
# Define the request payload
data = {
    "model": "claude-sonnet-4-5",  # latest Claude 3.5 model as of Oct 2024
    "max_tokens": 10000,
    "messages": [
        {"role": "user", "content": "Write a detailed protocol for creating this media with an Opentron Flex protocol. Include first the labware needed, solutions required in each labware and instructions on making them. Then secondly write as text the python script that i can run on an Opentron Flex robot. Minimize the number of tokens needed"}
    ]
}

# Send the POST request
response = requests.post(url, headers=headers, json=data)

# Parse and display the result
if response.status_code == 200:
    reply = response.json()
    print("Claude's response:\n")
    print(reply["content"][0]["text"])
else:
    print("Error:", response.status_code, response.text)
```
